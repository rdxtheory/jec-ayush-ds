```python
import pandas as pd
import numpy as np
```

```python
overall_stats = pd.read_csv('ml-100k/u.info', header=None)
print("Details of users, items and ratings involved in the loaded movielens dataset: ",list(overall_stats[0]))
```

```python

## same item id is same as movie id, item id column is renamed as movie id
column_names1 = ['user id','movie id','rating','timestamp']
dataset = pd.read_csv('ml-100k/u.data', sep='\t',header=None,names=column_names1)
dataset
```

```python

d = 'movie id | movie title | release date | video release date | IMDb URL | unknown | Action | Adventure | Animation | Children | Comedy | Crime | Documentary | Drama | Fantasy | Film-Noir | Horror | Musical | Mystery | Romance | Sci-Fi | Thriller | War | Western'
column_names2 = d.split(' | ')
column_names2
```

```python
items_dataset = pd.read_csv('ml-100k/u.item', sep='|',header=None,names=column_names2,encoding='latin-1')
items_dataset
```

```python

movie_dataset = items_dataset[['movie id','movie title']]
movie_dataset.head()
```

```python

#looking at length of original items_dataset and length of unique combination of rows in items_dataset after removing movie id column
len(items_dataset.groupby(by=column_names2[1:])),len(items_dataset)
```

```python
merged_dataset = pd.merge(dataset, movie_dataset, how='inner', on='movie id')
merged_dataset.head()
```

```python
merged_dataset[(merged_dataset['movie title'] == 'Chasing Amy (1997)') & (merged_dataset['user id'] == 894)]
```

```python
refined_dataset = merged_dataset.groupby(by=['user id','movie title'], as_index=False).agg({"rating":"mean"})

refined_dataset.head()
```

```python
user_to_movie_df=refined_dataset.pivot(index='user id',columns='movie title',values='rating').fillna(0)
```

```python
user_to_movie_df
```

```python
from scipy.sparse import csr_matrix
```

```python
user_to_movie_sparse_df=csr_matrix(user_to_movie_df.values)
```

```python
print(user_to_movie_sparse_df)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
knn_mod=NearestNeighbors(metric='cosine',algorithm='brute')
```

```python
knn_mod.fit(user_to_movie_sparse_df)
```

```python
#def a function that finds n simmilar users to the given inputed user
import numpy as np
def get_users(user,n=5):
  knn_inp=np.asarray([user_to_movie_df.values[user-1]]) #get the rating values of the concerned user
  dis,indices=knn_mod.kneighbors(knn_inp,n_neighbors=n+1) #get the distance and index of simmilar n users
  print("Top ",n," users who are simmilar to user ",user," are:")
  print(" ")
  for i in range(1,len(dis[0])):
    print(i," . User: ",indices[0][i]+1," seprated by a distance of ",dis[0][i])
  return indices.flatten()[1:] + 1,dis.flatten()[1:]


```

```python
user_id=int(input('enter the user whom to suggest movies(user id should be between 1-943):'))

simm_users,distance=get_users(user_id,15)

```

```python
simm_users
```

```python
distance
```

```python
weighted_list=distance/np.sum(distance)
```

```python
weighted_list.shape
```

```python
mov_ratings_for_simm_users=user_to_movie_df.values[simm_users]
```

```python
mov_ratings_for_simm_users.shape
```

```python
movie_list=user_to_movie_df.columns
movie_list.shape
```

```python
weighted_list=weighted_list[:,np.newaxis] + np.zeros(len(movie_list))
```

```python
weighted_list.shape
```

```python
new_weighted_ratings=weighted_list*mov_ratings_for_simm_users
```

```python
mean_rating_simm_users=new_weighted_ratings.sum(axis=0)
```

```python
mean_rating_simm_users
```

```python
def rec_movie(n):
  n=min(len(mean_rating_simm_users),n)
  print(list(movie_list[np.argsort(mean_rating_simm_users)[::-1][:n]]))
```
